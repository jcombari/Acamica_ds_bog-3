```python
sobreendeudamiento=[]
sobreendeudamiento=pd.DataFrame(sobreendeudamiento)
filenames=['Sal_consol_1605.xlsx', 'Sal_consol_1608.xlsx', 'Sal_consol_1611.xlsx', 'Sal_consol_1702.xlsx', 'Sal_consol_1705.xlsx', 'Sal_consol_1708.xlsx', 'Sal_consol_1711.xlsx', 'Sal_consol_1802.xlsx', 'Sal_consol_1805.xlsx', 'Sal_consol_1808.xlsx', 'Sal_consol_1811.xlsx', 'Sal_consol_1902.xlsx', 'Sal_consol_1905.xlsx', 'Sal_consol_1908.xlsx', 'Sal_consol_1911.xlsx']
os.chdir('C:\\Users\\NATALIA.GONZALEZ\\Desktop\\Hisotrico\\Consol')

for file in filenames:   
    pfile = pd.read_excel(file,sep=' ',dtype={'NUMERO_IDENTIFICACION': np.object})
    pfile['corte']=file[11:15]
    pfile.columns = pfile.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('@', '').str.replace('>', '').str.replace('á','a').str.replace('é','e').str.replace('í','i').str.replace('ó','o').str.replace('ú','u').str.replace(',','.')
    sobreendeudamiento=pd.concat([sobreendeudamiento, pfile] , axis=0)
```
